```python
import pandas as pd
import matplotlib.pyplot as plt
d=pd.read_csv("E:\python_DSBDA\heart.csv")
d
```

```python
d["chol"].fillna(d["chol"].median(),inplace=True)
```

```python
d
```

```python
x=d["trestbps"].mean()
d["trestbps"].fillna(x,inplace=True)
d
```

```python
d.dropna()
d
```

```python
y=d
y.drop_duplicates(inplace=True)
y
```

```python
for x in y.index:
    if y.loc[x,"age"]>70:
        y.drop(x,inplace=True)
        
```
